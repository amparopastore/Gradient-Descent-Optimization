# Gradient Descent for Function Minimization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Preliminaries

```python
# Given two functions f1(x,y)=(x-2)2 + (y-3)2, and f2(x,y)=(1 - (y-3))2 + 20((x+3) - (y-3)2)2, please implement gradient descent learning to search 
# for global minimum value for each function. 

def f1(x, y):
    return (x - 2)**2 + (y - 3)**2

def f2(x, y):
    return (1 - (y - 3))**2 + 20 * ((x + 3) - (y - 3)**2)**2

def grad_f1(x, y):
    df_dx = 2 * (x - 2)
    df_dy = 2 * (y - 3)
    return np.array([df_dx, df_dy])

def grad_f2(x, y):
    df_dx = 40 * ((x + 3) - (y - 3)**2)
    df_dy = -40 * (y - 3) * ((x + 3) - (y - 3)**2) - 2 * (1 - (y - 3))
    return np.array([df_dx, df_dy])

def gradient_descent (f, gradient, point, n, T):
    for i in range(T):
        grad = gradient(point[0], point[1])
        point = point - (n * grad)
        print(point)
    value = f(point[0], point[1])
    return point, value
```

### Plots

```python
# Use two Plots to show f1(x,y) and f2(x,y) values with respect to the two dimensional input x and y 
# (You can specify the range of the x and y values)

x_range = np.linspace(-10, 10, 100)
y_range = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x_range, y_range)

Z1 = f1(X, Y)
Z2 = f2(X, Y)

fig = plt.figure(figsize=(12, 6))

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, Z1, cmap='viridis')
ax1.set_title('f1(x, y)')

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(X, Y, Z2, cmap='viridis')
ax2.set_title('f2(x, y)')

plt.show()
```

### Implementation 
### learning rate = .5

```python
# Starting from initial value: (x,y)=(0,0), use learning rate h=0.5, report f1(x,y) and f2(x,y) values in T=100 iterations. 
# (your code can report f1(x,y) and f2(x,y) values as tables, or simple print out the values). 

initial_value = np.array([0, 0]) # initial value
n = .5 # learning rate
T = 100 # iterations

min_point_f1, min_value_f1 = gradient_descent(f1, grad_f1, initial_value, n, T)
min_point_f2, min_value_f2 = gradient_descent(f2, grad_f2, initial_value, n, T)
```

```python
print("Minimum value for f1:", min_value_f1, "at point:", min_point_f1)
print("Minimum value for f2:", min_value_f2, "at point:", min_point_f2)
```

**Was it effective?**

In the first attempt to use gradient descent to find a global minimum value for the functions, the learning rate of 0.5 caused the gradient calculations for `f2` to become extremely large creating numerical instability. This indicates that the learning rate is too high and is leading to large updates that push the parameters into regions with very high gradients. 

Now, we'll adjust the learning rate to 0.01.

### learning rate = .01

```python
# Following step 2, please change your code (e.g., using different learning rates, such as h=0.01) to try to search minimum for f1(x,y) and f2(x,y), 
# respectively. Run algorithms for T=100 iterations Explain the motivation of your changes, and the final minimum values 
n_2 = 0.01
min_point_f1, min_value_f1 = gradient_descent(f1, grad_f1, initial_value, n_2, T)
min_point_f2, min_value_f2 = gradient_descent(f2, grad_f2, initial_value, n_2, T)
```

```python
print("Minimum value for f1:", min_value_f1, "at point:", min_point_f1)
print("Minimum value for f2:", min_value_f2, "at point:", min_point_f2)
```

This learning rate worked better for f1, but it is still too large for f2, so let's try a smaller one.

### learning rate = .001

```python
n_3 = 0.001
min_point_f1, min_value_f1 = gradient_descent(f1, grad_f1, initial_value, n_3, T)
min_point_f2, min_value_f2 = gradient_descent(f2, grad_f2, initial_value, n_3, T)
```

```python
print("Final minimum value for f1:", min_value_f1, "at point:", min_point_f1)
print("Final minimum value for f2:", min_value_f2, "at point:", min_point_f2)
```
